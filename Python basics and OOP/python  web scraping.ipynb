```python
#web scraping is nothing but the,obtaining info from the net,downloading the images from net etc.
# we use 3 libraries,requests,lxml,beautiful soup,
#some web pages do not support web scraping,so first we should,request them,later we should proceed...
```

```python
import requests
res = requests.get("https://en.wikipedia.org/wiki/Pawan_Kalyan")
#requests module has method get ,syntax is requests.get(),which does the work of requesting the web page,some pages do not support
#so in that case it would give error,if doent give error,it mean requested successfully
```

```python
type(res) #it is a special one
```

```python
res.text #text is an attribute,which gives the html & css text of the web page in single gaint string,inorder to get
#in actual html format we should use the bs4 module..
```

```python
import bs4 #(beautiful soup allows us to easily grab the info)
soup = bs4.BeautifulSoup(res.text,'lxml') #note it is case sensetive
#it takes two arguments ,1st jaeint html string,and lxml, bs4 would go through the string and figuers out ,which is html tag,
#css id ,css class.and returns the text in actual html &css text
```

```python
#now if we print soup,we would get actual text in html &css format
soup
```

GRABBING THE TITLE OF WEB PAGE..

```python
#we can get by using the selct methid.
soup.select('title') #selsect func searches for the title tag,and by default returns the list,because there may be many titles..
#in our case titlte is only pavan kalyan - Wikipedia ,we can find title ,it is embedded inside title tags..
```

GRABBING THE ELEMENTS ASSOCIATED WITH  A CLASS 

```python
#notations 
#SYNTAX                                               MATCH RESULTS
#soup.select('div')                     all elements the 'div' tag , like p,h1,title tags
#soup.select('.some_class)        Elements containing class = 'some_class (class are in CSS)you can find id 's in html text 
#soup.select('#some_id')          Elements containing id = 'some_id' (id are in also CSS)you can find id 's in html text    
#soup.select('div span')          Any elements named span within a div element
#soup.select('div>span')          Any elements named span directly within a div elementwith nothing in between
```

```python
soup.select('.left-aligned')#list of text in the class tag left-aligned.
#left-aligned is the class name
#note .(dot) this is symbol of donating class,this is called period in html languagge.
# we use #for id's
```

```python
# we can go through them in order to get the text in class
for item in soup.select('.left-aligned') :
    print(item.getText(),'\n')
```

```python
soup.select('.toctext') #if we donot use .,it would treat toctext as some tag..which would give error ,because it is not a tag 
```

```python
for item in soup.select('.toctext') :
    print(item.getText()) #getText() method is used to get atual text,without html tags..
```

```python
#DOWNLOADING THE IMAGE..
image_link = requests.get("https://upload.wikimedia.org/wikipedia/commons/thumb/2/2a/Telugu_actor_Pawan_Kalyan_meets_Narendra_Modi.jpg/220px-Telugu_actor_Pawan_Kalyan_meets_Narendra_Modi.jpg") 
#requesting the image source url page..
#make sure that you add https: before the image url,if not it would give error.. 
```

```python
image_link.content
```

```python
#above is the raw _content of the image file..it is actually a binary code..pyhton can understand it..
#we gonna save image as file in present working directory.
```

```python
file = open('pavan_modi_photo','wb') #we open in wb (write binary ) mode because it is actually a binary code file..
```

```python
file.write(image_link.content)
```

```python
file.close()
```
